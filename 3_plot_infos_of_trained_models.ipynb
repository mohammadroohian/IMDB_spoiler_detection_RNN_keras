```python
from os.path import join
from os import listdir

import _utils.helper as hlp

def show_model_info(model_h):
    for i, v in model_h['hyper_parameters'].items():
        val = v
        if type(val) == str:
            if 'RMSprop' in v: val = 'RMSprop'
            elif 'adam' in v: val = 'adam'
        print(i+':', '\t',val)
    duration = model_h.get('train_duration')
    if duration != None: duration = round(duration/60, 1)
    #print(model_h['history'].model.get_config())
    print('train duration:\t', duration)
    model_h['history'].model.summary()
```

```python
#from keras import layers
#from keras import models

models_path = './models/samples_300k'
models_list = []
#models_list = listdir(models_path)
models_list.append('15805_0_2_1_48_Conv1D_CuDNNLSTM')
models_list.append('15805_0_4_9_65_Conv1D_CuDNNLSTM')
models_list.append('15805_0_7_8_46_Conv1D_CuDNNLSTM')
models_list.append('15805_0_9_6_81_Conv1D_CuDNNLSTM_best_of_all')
models_list.append('15805_1_2_0_58_Conv1D_CuDNNLSTM')
models_list.append('15805_1_3_4_32_Conv1D_CuDNNLSTM')


#models_list.remove('old')
#models_list = [models_list[0]]
for i in models_list: 
    model_h = hlp.load(join(models_path, i))
    print('model file name is:', i)
    hlp.show_model_plot(model_h, (.95,.02), (.95,1.1));
    show_model_info(model_h)
    print('-'.join(['' for i in range(0, 16)]))
```
